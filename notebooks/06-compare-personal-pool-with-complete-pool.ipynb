```python
from pymongo import MongoClient
```

```python
uri = "mongodb://localhost:27017"
db_name = "mtg_decks_db"
client = MongoClient(uri)
db = client[db_name]
```

```python
pipeline = [
    {"$lookup": {"from": "personal_pool", "localField": "name",
                 "foreignField": "name", "as": "personal_data"}},
    {"$project": {"name": 1, "scryfall_name": 1, "count": 1,
                  "personal_count": {"$sum": "$personal_data.count"}}},
    {"$project": {"_id": 0, "name": 1, "scryfall_name": 1,
                  "missing": {"$subtract": ["$count", "$personal_count"]}}},
    {"$match": {"missing": {"$gt": 0}}},
    {"$sort": {"scryfall_name": 1}},
]
```

```python
cards = db.pool.aggregate(pipeline)

for card in cards:
    print(f"{card["missing"]} {card["scryfall_name"]}")
```
